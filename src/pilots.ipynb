## Cấu trúc chèn pilot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_slots = 15
n_subs = 15

fig, axs = plt.subplots(1, 2, figsize=(6, 3))

# 1. Block-type: VD Chèn pilot vào mỗi 4 khe thời gian
time_spacing = 4
freq_spacing = 1
slots = np.arange(0, n_slots, time_spacing)
subs = np.arange(0, n_subs, freq_spacing)
T, F = np.meshgrid(slots, subs)
axs[0].scatter(T, F, label='Vị trí pilot')
axs[0].legend()
axs[0].set_title('Block-Type')

# 2. Comb-type: VD Chèn pilot vào mỗi 8 sóng mang con
time_spacing = 1
freq_spacing = 4
slots = np.arange(0, n_slots, time_spacing)
subs = np.arange(0, n_subs, freq_spacing)
T, F = np.meshgrid(slots, subs)
axs[1].scatter(T, F, label='Vị trí pilot')
axs[1].legend()
axs[1].set_title('Comb-Type')

for ax in axs:
    ax.set_xlabel('Khe thời gian')
    ax.set_ylabel('Sóng mang con')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
n_slots = 15
n_subs = 15

fig, ax = plt.subplots(figsize=(3, 3))

# 3. Lattice-type: VD Chèn pilot vào mỗi 4 khe thời gian 
# và mỗi 4 sóng mang con 
time_spacing = 4
freq_spacing = 4
slots = np.arange(0, n_slots, time_spacing)
subs = np.arange(0, n_subs, freq_spacing)
T, F = np.meshgrid(slots, subs)
ax.scatter(T, F, label='Vị trí pilot')
ax.legend()
ax.set_title('Lattice-Type')
ax.set_xlabel('Khe thời gian')
ax.set_ylabel('Sóng mang con')
ax.grid(True)

plt.tight_layout()
plt.show()
```
